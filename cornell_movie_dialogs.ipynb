```python
import json
import pandas as pd
import numpy as np

from IPython.display import display
```

```python
# some useful code to deal with the crap data from Cornell Uni

def read_cornell(path, *, header=None, engine='python', sep=r' \+\+\+\$\+\+\+ ', **kwargs):
    "load from a very non-standard CSV file"
    return pd.read_csv(path, header=header, engine=engine, sep=sep, **kwargs)

def loads_almost_json(s):
    "json.loads when you've been given a json object with invalid quotes, pretty fragile!"
    return json.loads(s.replace("'", '"'))
```

```python
mov_movies = read_cornell(
    'data/cornell_movie_dialogs_corpus/movie_titles_metadata.txt',
    names=['movie', 'title', 'year', 'imdb_rating', 'imdb_votes', 'genres'],
    index_col=0)

mov_movies.genres = mov_movies.genres.apply(loads_almost_json)

mov_movies.head(12)
```

```python
mov_conv = read_cornell(
    'data/cornell_movie_dialogs_corpus/movie_conversations.txt',
    names=['char1', 'char2', 'movie', 'lines'])

mov_conv.lines = mov_conv.lines.apply(loads_almost_json)

mov_conv.head(12)
```

```python
mov_lines = read_cornell(
    'data/cornell_movie_dialogs_corpus/movie_lines.txt',
    names=['line', 'char', 'movie', 'char_name', 'text'],
    index_col=0)

mov_lines.head(12)
```
